# Data Cleaning

```python
# Mental Health Dataset

# Import necessary modules
import pandas as pd

print("Please download the Mental Health CSV and store it locally.")
print("It can be found for download at the following link: https://www.kaggle.com/datasets/imtkaggleteam/mental-health?select=1-+mental-illnesses-prevalence.csv")

# Read CSV
mental_health_df = pd.read_csv('1- mental-illnesses-prevalence.csv', header = 0, encoding='latin1')

# Clean up column names
mental_health_df.rename(
   columns={"Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized": "Schizophrenia Disorders",
            "Depressive disorders (share of population) - Sex: Both - Age: Age-standardized": "Depressive Disorders",
            "Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized": "Anxiety Disorders",
            "Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized": "Bipolar Disorders",
            "Eating disorders (share of population) - Sex: Both - Age: Age-standardized": "Eating Disorders",
            "Entity": "Country/Territory"
           }, inplace=True
)

# Only keep data from the US
mental_health_us_df = mental_health_df[(mental_health_df['Country/Territory'] == "United States")]

mental_health_us_df
# len(mental_health_df)
```

```python
# Causes of death dataset

print("Please download the Causes of Death CSV and store it locally.")
print("It can be found for download at the following link: https://www.kaggle.com/datasets/iamsouravbanerjee/cause-of-deaths-around-the-world")

# Read CSV
causes_of_death_df = pd.read_csv('cause_of_deaths.csv', header = 0, encoding='latin1')

# Only keep data from the US
causes_of_death_us_df = causes_of_death_df[(causes_of_death_df['Country/Territory'] == "United States")]

causes_of_death_us_df
```

```python
# Join the dataframes on country/territory, year, and code, while dropping any rows with missing data
merged_df = mental_health_us_df.merge(causes_of_death_us_df, on = ["Country/Territory", "Year", "Code"], how = "outer").dropna()

merged_df
```

```python
# Add US population by year
# Population data source: https://www.multpl.com/united-states-population/table/by-year
merged_df.insert(0, "Population", [251000000, 254530000, 258060000, 261350000, 264510000, 267670000, 270860000, 274110000, 277270000, 280470000, 283440000, 286140000, 288800000, 291360000, 294020000, 296760000, 299660000, 302530000, 305300000, 307950000, 310350000, 312590000, 314870000, 317060000, 319410000, 321710000, 323910000, 325800000, 327430000, 329020000])
merged_df
```

```python
# Multiply population column by disorder columns (divided by 100 to convert to percentage)

merged_df['Number of Schizophrenia Disorders'] = ((merged_df['Schizophrenia Disorders'])/100) * merged_df['Population']
merged_df['Number of Depressive Disorders'] = (merged_df['Depressive Disorders']/100) * merged_df['Population']
merged_df['Number of Anxiety Disorders'] = (merged_df['Anxiety Disorders']/100) * merged_df['Population']
merged_df['Number of Bipolar Disorders'] = (merged_df['Bipolar Disorders']/100) * merged_df['Population']
merged_df['Number of Eating Disorders'] = (merged_df['Eating Disorders']/100) * merged_df['Population']

merged_df
```

```python
# Visualization Necessary Imports
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = merged_df["Year"]
y = merged_df["Cardiovascular Diseases"]
x2 = x = merged_df["Year"]
y2 = merged_df["Number of Eating Disorders"]
# Create a basic line plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Cardiovascular Disease Deaths')
plt.plot(x2, y2, label = 'Cases of Eating Disorders')
plt.title('Cases of Eating Disorders and Cardiovascular Disease Deaths from 1990-2019')
plt.xlabel('Year')
plt.ylabel('Total Cases/Deaths')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

selected_columns = ['Depressive Disorders','Chronic Kidney Disease', 'Cirrhosis and Other Chronic Liver Diseases', 'Alcohol Use Disorders', 'Road Injuries', 'Drug Use Disorders', 'Digestive Diseases']
specific_df = merged_df[selected_columns]

# Calculate the correlation matrix for the selected columns
correlation_matrix = specific_df.corr()

sns.heatmap(
    correlation_matrix, annot=True,fmt=".2f", cmap="YlGnBu", linewidths=0.5,cbar_kws={"shrink": 0.8, "label": "Correlation Coefficient"})
plt.title("Correlations of Death Causes and Mental Disorders", fontsize=16, pad=20, weight="bold")
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks(rotation=0, fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Plot prevalence of selected disorders over time
import matplotlib.pyplot as plt
import seaborn as sns

selected_disorders = ['Nutritional Deficiencies','Protein-Energy Malnutrition', 'Neonatal Disorders', 'Maternal Disorders']
df_melted = merged_df.melt(id_vars=['Year'], value_vars=selected_disorders, var_name='Deficiency', value_name='Prevalence')

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_melted, x='Year', y='Prevalence', hue='Deficiency')
plt.title('Cases of Health Deficiencies Over Time')
plt.xlabel('Year')
plt.ylabel('Cases')
plt.legend(title='Deficiency')
plt.show()
```
